```python
import pandas as pd
import os

FILE_NAME = "Gyminsights_dataset2.csv"


possible_paths = [
    FILE_NAME,  
    os.path.join("jupyter_notebooks", FILE_NAME),  
]


FULL_PATH = None
for path in possible_paths:
    if os.path.exists(path):
        FULL_PATH = path
        break

if FULL_PATH is None:
    raise FileNotFoundError(f"File '{FILE_NAME}' not found in expected locations: {possible_paths}")


df = pd.read_csv(FULL_PATH)

df.head()
```
